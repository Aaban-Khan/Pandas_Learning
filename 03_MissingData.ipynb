```python
import pandas as pd 
import numpy as np 
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

Filling Missing Data

```python
df.fillna(0)
```

```python
values={"A":100, "B":400,"C":200,"D":900}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
